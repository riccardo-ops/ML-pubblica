## 3.5 Apply che modifica i valori nel dataframe

```python
# situazione : volgio iterare sulla colonna Churn e voglio mmodificare i valori booleani in valori binari
# devo quindi filtrare e poi applicare la maschera al dataset
# creo una nuova colonna e applico una funzione lambda che fa esattamente questo
df['Churn Binario'] = df['Churn'].apply(lambda x : 1 if x == True else 0)
```
